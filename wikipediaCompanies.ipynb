```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from os import remove
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"
page = requests.get(url)

soup = BeautifulSoup(page.text, features="html.parser")
```

```python
data = soup.find('tbody')
dataList = []
headersList = []
i = 0
for row in data:
    colList = []
    for col in row:          
        if type(soup.find('td')) == type(col):
          if type(col.get('class')) != type(None) and 'table' in col.get('class')[0]:
            colList.append(col.get('data-sort-value')) 
          text = col.text
          text = " ".join(text.split())
          text = text.replace('$', '').replace(",", "")
          if text != '' and '[' not in text:
            colList.append(int(text) if text.isnumeric() else (float(text) if text.replace(".", "").isnumeric() else text))
    if i == 0:
      headersList = colList
      headersList.remove('Ref.')
      headersList.remove(None)
      headersList.insert(6, "Headquarters")
    if len(colList) > 0 and i > 4:
      dataList.append(colList)
    i+=2
      
df = pd.DataFrame(dataList, columns=headersList)
df.set_index('Rank', inplace=True)
df
```
